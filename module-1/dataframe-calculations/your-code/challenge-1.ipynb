#### Create a new column called `Combo Type` whose value combines `Type 1` and `Type 2`.

Rules:

* If both `Type 1` and `Type 2` have valid values, the `Combo Type` value should contain both values in the form of `<Type 1> <Type 2>`. For example, if `Type 1` value is `Grass` and `Type 2` value is `Poison`, `Combo Type` will be `Grass-Poison`.

* If `Type 1` has valid value but `Type 2` is not, `Combo Type` will be the same as `Type 1`. For example, if `Type 1` is `Fire` whereas `Type 2` is `NaN`, `Combo Type` will be `Fire`.

```python
# your code here

#Function to create a new column with types of pokemon
def combo(type1,type2):
    if pd.isna(type1)==False and pd.isna(type2)==False:   #If type1 and type2 aren't equal to NaN, return 'type1-type2'
        combo = type1 + '-' + type2
        #print('Entra 1: ', combo)
    elif pd.isna(type1):   #If type1 is NaN, return 'type2'
        combo = type2
        #print('Entra 2: ', combo)
    elif pd.isna(type2):   #If type2 is NaN, return 'type1'
        combo = type1
        #print('Entra 3: ', combo)
    return combo
```

```python
# Apply the function below to the dataframe colunm

#I use try-except to test the loop
try:
    #I use the lambda function to be able to perform the required operations with two columns of the same dataframe. And 
    #after that, apply this function to pokemon_df
    pokemon_df['Combo Type'] = pokemon_df.apply(lambda x: combo(x['Type 1'], x['Type 2']), axis=1)
except:
    print('Error: ', df['Name'])
```
